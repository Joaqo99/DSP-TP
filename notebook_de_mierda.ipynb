```python
import numpy as np
import pandas as pd
import time
import audio_functions as auf
```

```python
t60_list = np.linspace(0.2, 0.8, 10)

dmic_list = np.linspace(0.1, 1, 5)

source_x_list = np.random.normal(5, 2, 5)
source_y_list = np.random.normal(5, 2, 5)
source_list = []

for i in range(5):
    x_pos = np.round(source_x_list[i], 1)
    y_pos = np.round(source_y_list[i], 1)
    source_pos = [x_pos, y_pos, 1]
    source_list.append(source_pos)

mic_x_list = np.random.normal(5, 2, 5)
mic_y_list = np.random.normal(5, 2, 5)
mic_list = []

for i in range(5):
    x_pos = np.round(mic_x_list[i], 1)
    y_pos = np.round(mic_y_list[i], 1)
    mic_pos = [x_pos, y_pos, 1]
    mic_list.append(mic_pos)
```

```python
print(mic_list)
```

```python
all_dfs = []
t60 = t60_list[0]
#for t60 in t60_list:
for d in dmic_list:
    for s_pos in source_list:
        for m_pos in mic_list:
            sim_name = f"sim_T{t60:.2f}s_d{d:.1f}_s[{s_pos[0]:.1f},{s_pos[1]:.1f},{s_pos[2]:.1f}]_m[{m_pos[0]:.1f},{m_pos[1]:.1f},{m_pos[2]:.1f}]"

            mods = [
                {"var":"room", "param":'dim', "value":[10, 10, 3]},
                {"var":"room",      "param":"t60",      "value": float(t60)},
                {"var":"mic_array", "param":"d",        "value": float(d)},
                {"var":"mic_array", "param":"position", "value":s_pos},
                {"var":"source",    "param":"position", "value":m_pos},]

            auf.gen_simulation_dict(sim_name, *mods)
            
            df = auf.process_simulation_data(sim_name)
            df["t60"]      = t60
            df["d"]        = d

            all_dfs.append(df)
```
