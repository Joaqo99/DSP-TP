```python
import numpy as np
import pandas as pd
import time
import audio_functions as auf
```

```python
t60_list = np.linspace(0.2, 1.1, 10)

dmic_list = [0.1]


source_x_list = np.random.randint(10, size=10)
source_y_list = np.random.randint(10, size=10)
source_list = []

for i in range(10):
    x_pos = int(source_x_list[i])
    y_pos = int(source_y_list[i])
    source_pos = [x_pos, y_pos, 1]
    source_list.append(source_pos)

mic_x_list = np.random.randint(4, high=11, size=10)
mic_y_list = np.random.randint(0, high=11, size=10)
mic_list = []

for i in range(10):
    x_pos = int(mic_x_list[i])
    y_pos = int(mic_y_list[i])
    mic_pos = [x_pos, y_pos, 1]
    mic_list.append(mic_pos)
```

```python
print(t60_list)
```

```python
names = []

for t60 in t60_list:
    for d in dmic_list:
        for s_pos in source_list:
            for m_pos in mic_list:
                sim_name = f"sim_T{t60:.2f}s_d{d:.1f}_s[{s_pos[0]:.1f},{s_pos[1]:.1f},{s_pos[2]:.1f}]_m[{m_pos[0]:.1f},{m_pos[1]:.1f},{m_pos[2]:.1f}]"

                mods = [
                    {"var":"room", "param":'dim', "value":[10, 10, 3]},
                    {"var":"room",      "param":"t60",      "value": float(t60)},
                    {"var":"mic_array", "param":"d",        "value": float(d)},
                    {"var":"mic_array", "param":"position", "value":m_pos},
                    {"var":"source",    "param":"position", "value":s_pos}]

                auf.gen_simulation_dict(sim_name, *mods)

                names.append(sim_name)
            
```

```python
df = auf.process_simulation_data(*names, df_values= ["room.t60", "mic_array.d", "mic_array.position", "source.position"])
df.to_csv("simulación_rt60 - correjido.csv")
```
