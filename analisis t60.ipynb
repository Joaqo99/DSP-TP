```python
import plot
import pandas as pd
```

```python
df = pd.read_csv("simulación_rt60 - correjido.csv")
df = df.drop("Unnamed: 0", axis=1)
df['room.t60'] = df['room.t60'].round(2)
df
```

```python
t60_bars = lambda: plot.plot_bars_error(df, "room.t60", "error", errorbar="sd", grid=True, rotation=45, y_label="Valor medio del error angular", x_label="T60 [s]", show=False)
t60_w_lines = lambda: plot.plot_lines_error(df, "room.t60", "error", hue="method", grid=True, rotation=45, y_label="Valor medio del error angular", x_label="T60 [s]", legend=True, show=False)

plot.multiplot(t60_bars, t60_w_lines, figsize=(6, 7), ncols=1)
```
